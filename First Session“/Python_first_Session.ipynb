```python
DT = {}
for a in DP:
    if isinstance(DP[a], dict):
        print('yes')
        keys = list(DP[a].keys())
        DP2 = DP[a][keys[-1]]
        DP2_keys = list(DP2.keys())
        DT.update({DP2_keys[0]:DP2[DP2_keys[0]]})
    else:
        DT.update({a:DP[a][0]})
```
