# The Water Cycle: Nature's Continuous Process

The water cycle, also known as the hydrological cycle, is a fundamental environmental process that describes how water moves through the Earth and its atmosphere. This continuous cycle is crucial for sustaining life, regulating climate, and maintaining ecosystems.

## Key Stages of the Water Cycle

1. **Evaporation**:
   Water from oceans, rivers, lakes, and even soil evaporates into the atmosphere due to heat from the sun. This process transforms liquid water into water vapor, a gas that rises into the air.

2. **Condensation**:
   As water vapor rises, it cools and condenses to form tiny droplets, creating clouds. This stage is essential for the formation of precipitation, as it gathers moisture in the atmosphere.

3. **Precipitation**:
   Eventually, the droplets in clouds combine to form larger droplets. When they become heavy enough, they fall back to Earth as precipitation, which can be rain, snow, sleet, or hail, depending on atmospheric conditions.

4. **Collection**:
   Once the water reaches the ground, it collects in bodies of water like rivers, lakes, and oceans. Some of it also infiltrates the soil, replenishing groundwater supplies. This collected water will eventually evaporate again, continuing the cycle.

## Importance of the Water Cycle

- **Sustaining Ecosystems**: The water cycle is essential for all living organisms. It provides fresh water for drinking, agriculture, and sanitation.
- **Climate Regulation**: Water vapor plays a significant role in regulating temperature and weather patterns. It helps to stabilize the Earth's climate by absorbing and redistributing heat.
- **Nutrient Distribution**: As water moves through the cycle, it helps in the transport of nutrients, minerals, and sediments, supporting various ecosystems.

## Human Impact on the Water Cycle

Human activities, such as deforestation, urbanization, and pollution, can disrupt the natural water cycle. For example, urban areas often lead to increased runoff and decreased groundwater recharge. Climate change is also affecting precipitation patterns and the availability of fresh water.

# Understanding Drought and Its Effects on Agriculture

## What is Drought?

Drought is a prolonged period of abnormally low rainfall, leading to a water shortage. It can occur in various forms, including meteorological, agricultural, hydrological, and socio-economic droughts. Each type impacts agriculture differently:

- **Meteorological Drought**: A significant reduction in precipitation over a specific area.
- **Agricultural Drought**: A deficiency in soil moisture that hampers crop growth.
- **Hydrological Drought**: Decreased water levels in rivers, lakes, and reservoirs.
- **Socio-economic Drought**: Impacts on the economy and society due to water shortages.

## Causes of Drought

Drought can result from various factors, including:

- **Climate Change**: Altered weather patterns can lead to prolonged dry periods.
- **Deforestation**: Reduces moisture in the atmosphere, affecting rainfall.
- **Poor Land Management**: Overuse of water resources can deplete aquifers.
- **Natural Variability**: Some regions naturally experience cycles of dry and wet conditions.

## Effects of Drought on Agriculture

1. **Reduced Crop Yields**:
   - Insufficient water leads to lower crop productivity.
   - Some plants may fail to germinate or mature.

2. **Soil Degradation**:
   - Drought can cause soil erosion and loss of fertility.
   - Lack of moisture can lead to hard, crusty soil that is difficult to cultivate.

3. **Water Shortages**:
   - Farmers may face restrictions on water usage.
   - Irrigation systems may become less effective, impacting crop irrigation.

4. **Economic Impact**:
   - Decreased crop yields can lead to financial hardship for farmers.
   - Higher food prices may result from reduced supply.

5. **Increased Pest and Disease Pressure**:
   - Stressed plants are more susceptible to pests and diseases.
   - Drought conditions may favor certain pests or pathogens.


## Adaptation Strategies

Farmers and agricultural communities can implement several strategies to mitigate the effects of drought:

1. **Water Conservation Techniques**:
   - Rainwater harvesting and efficient irrigation systems (e.g., drip irrigation).
   - Mulching to retain soil moisture.

2. **Crop Selection**:
   - Choosing drought-resistant crop varieties that require less water.
   - Rotating crops to maintain soil health.

3. **Soil Management**:
   - Implementing practices like cover cropping and reduced tillage to improve soil structure and moisture retention.

4. **Agroforestry**:
   - Integrating trees and shrubs into agricultural systems to enhance biodiversity and water retention.

5. **Education and Awareness**:
   - Training farmers on drought management techniques and sustainable practices.


# Agro-Hydrology Indices Computation and ERA5 Data Download

This notebook calculates drought, aridity, and crop-related indices based on ERA5 climate data. We'll use `cubo` for data access and `pyet` for potential evapotranspiration calculations.

## Key Indices and Equations
### 1. Drought Indices
**Standardized Precipitation Index (SPI):**
SPI is calculated using precipitation data over a selected period (monthly/seasonal).
\[ \text{SPI} = \frac{P_i - \overline{P}}{\sigma_P} \]

**Palmer Drought Severity Index (PDSI):**
PDSI requires soil moisture and temperature.

### 2. Aridity Indices
**De Martonne Aridity Index (AI):**
\[ \text{AI} = \frac{P}{T + 10} \]
where:
- \(P\): Mean annual precipitation (mm)
- \(T\): Mean annual temperature (°C)

### 3. Crop-Related Indices
**Crop Water Stress Index (CWSI):**
\[ \text{CWSI} = \frac{ET_p - ET_a}{ET_p} \]
- \(ET_p\): Potential evapotranspiration
- \(ET_a\): Actual evapotranspiration

# What is Evapotranspiration?

Evapotranspiration (ET) is the combined process of water vapor moving from the soil and plants into the atmosphere. It encompasses two key components: **evaporation** and **transpiration**. This process is essential for maintaining the water cycle, influencing weather patterns, and supporting plant health.

### Components of Evapotranspiration

1. **Evaporation**  
   - **Definition**: The process whereby liquid water changes into vapor and enters the atmosphere.  
   - **Sources**:  
     - **Soil Surface**: Water evaporates directly from the ground, particularly in areas with bare soil.  
     - **Water Bodies**: Lakes, rivers, and reservoirs contribute significantly to evaporation.  
     - **Atmospheric Factors**: Temperature, humidity, wind speed, and solar radiation all influence evaporation rates. Higher temperatures and wind speeds typically increase evaporation.

2. **Transpiration**  
   - **Definition**: The process by which water absorbed by roots is released as vapor through small pores (stomata) in leaves.  
   - **Factors Influencing Transpiration**:  
     - **Plant Species**: Different plants have varying transpiration rates based on their physiology.  
     - **Soil Moisture**: Availability of water in the soil directly affects transpiration. Plants will reduce transpiration during drought conditions.  
     - **Environmental Conditions**: Light intensity, temperature, and humidity play crucial roles. For example, higher temperatures and greater solar radiation increase transpiration rates.

### Importance of Evapotranspiration

- **Water Balance**: ET is critical for calculating the water balance in ecosystems. It helps determine how much water is available for runoff, groundwater recharge, and plant use.  
- **Climate Regulation**: ET influences local climates by affecting humidity and temperature. Areas with high ET may experience cooler temperatures due to the cooling effect of water vapor.  
- **Agricultural Practices**: Understanding ET is vital for managing irrigation, optimizing crop yields, and ensuring sustainable water use.  
- **Ecosystem Health**: ET contributes to maintaining soil moisture and supporting plant health, which is essential for biodiversity.

## Measuring Evapotranspiration

Various techniques exist to measure or estimate ET, each with its advantages and limitations.

1. **Direct Measurement**  
   - **Lysimeters**: These devices measure the water loss from a soil column by collecting and weighing the water that drains through it. They provide accurate measurements but can be expensive and labor-intensive to maintain.

2. **Meteorological Methods**  
   - **Penman-Monteith Equation**: This widely used method calculates ET based on energy balance and aerodynamic principles. It requires data on temperature, humidity, wind speed, and solar radiation.  
   - **Hargreaves Equation**: A simpler method that uses temperature data (maximum and minimum) to estimate potential ET. It is less data-intensive but less accurate under certain conditions.

3. **Remote Sensing**: Satellite and aerial imagery can be used to estimate ET over larger areas by analyzing vegetation indices and surface temperature.

4. **Empirical Models**: These models estimate ET based on climatic and soil data, often calibrated for specific regions or crops. Examples include the FAO-56 model, which uses reference ET to estimate crop-specific ET.

## Actual vs. Potential Evapotranspiration

### Actual Evapotranspiration (AET)

- **Definition**: AET represents the amount of water actually evaporated and transpired under existing environmental conditions.  
- **Measurement Factors**:  
  - **Soil Moisture Availability**: AET is limited by the water present in the soil. If soil moisture is low, AET decreases.  
  - **Plant Health**: Healthy, well-watered plants transpire more than stressed plants.  
  - **Weather Conditions**: Variations in temperature, humidity, and wind can significantly impact AET levels.

### Potential Evapotranspiration (PET)

- **Definition**: PET is the maximum amount of ET that could occur if sufficient moisture were available in the soil.  
- **Determining Factors**:  
  - **Climate Variables**: PET is influenced primarily by temperature, solar radiation, and atmospheric pressure.  
  - **Vegetation Type**: Different plants have varying transpiration potentials, affecting PET estimates.

### Key Differences

- **Dependency**: AET is constrained by moisture availability, while PET assumes unlimited water.  
- **Application**: Understanding both AET and PET is crucial for water management, especially in agriculture, where irrigation decisions are made based on these values.

## Grass Evapotranspiration

Grasslands are significant ecosystems that contribute to global biodiversity and serve as important agricultural lands. The ET from grass is influenced by multiple factors:

### Factors Affecting Grass Evapotranspiration

1. **Grass Height and Type**: Taller grasses typically transpire more due to greater leaf area, which facilitates more stomatal openings for water vapor release.  
2. **Soil Moisture**: The availability of water in the soil directly impacts the transpiration rates of grass. Drought conditions can significantly reduce ET.  
3. **Weather Conditions**: Temperature, wind speed, and sunlight affect both evaporation from the soil and transpiration from the grass. For example, higher temperatures increase evaporation rates.  
4. **Growth Stages**: Grass in different growth stages (e.g., germination, vegetative, flowering) exhibits varying ET rates. During peak growth, ET is generally higher.

### Estimating Grass Evapotranspiration

1. **Reference Evapotranspiration (ETo)**: This standardized measure serves as a baseline for estimating ET for various crops, including grass. ETo can be calculated using meteorological data.  
2. **Crop Coefficients (Kc)**: These coefficients adjust ETo values to estimate the actual ET for grass, factoring in the specific growth stage and health of the grass.

### Practical Applications

- **Irrigation Management**: Understanding grass ET helps farmers and landscapers optimize irrigation schedules, ensuring plants receive adequate water without waste.  
- **Ecological Studies**: Monitoring grass ET can provide insights into ecosystem health and water use efficiency, aiding in conservation efforts.

## Compute Potential Evapotranspiration Using `pyet`

```python
lat, lon = 51.344661, 9.844198
da = cubo.create(
    lat=lat, # Central latitude of the cube
    lon=lon, # Central longitude of the cube
    collection="ECMWF/ERA5/DAILY", # Id of the GEE collection
    bands=["mean_2m_air_temperature","minimum_2m_air_temperature","maximum_2m_air_temperature"], # Bands to retrieve
    start_date="2016-01-01", # Start date of the cube
    end_date="2016-12-31", # End date of the cube
    edge_size=100, # Edge size of the cube (px)
    resolution=11000, # Pixel size of the cube (m)
    gee=True # Use GEE instead of STAC
)
```

```python
import pyet as pet

# Example climate data for PET calculation
mean_temperature = da.sel(band='mean_2m_air_temperature')  # °C
max_temperature = da.sel(band='maximum_2m_air_temperature')  # °C
min_temperature = da.sel(band='minimum_2m_air_temperature')  # °C
# Calculate PET using the Penman-Monteith method
pet_values = pet.hargreaves(
    tmean = mean_temperature,
    tmax  = max_temperature,
    tmin  = min_temperature,
    lat   = lat

)
print("Potential Evapotranspiration:", pet_values)
```

```python
pet_values.mean('x').mean('y').plot()
```
