# defining a function

```python
def divison_appending(length):
    '''
    ########### This function is dividing values based on their division
    ########### to the 2 and 5

    lenght: number, number of values to be appended
    output: list of values
    '''
    # empty list
    L = []
    # for loop for appending
    for x in range(0,length,1):
        # if condition for testing the division to 2 and 5
        if x%2==0 or x%5==0:
            L.append(x)
        # if condition for testing the division to 3
        elif x%3==0:
            L.append(1/x)
        else:
            L.append(2/x)
    return L
```

```python
L = divison_appending(10)
print(L)
```

## isinstance is a built in function
https://docs.python.org/3/library/functions.html#isinstance

```python
L = [0,0,10,2,3,4,5,6]
A = {}
for i in range(len(L)):
    try:
        print(1/L[i+1])
    except Exception as e:  # Catch all exceptions
        if isinstance(e, ZeroDivisionError):
            A.update({'ZeroDivisionError':L[i]})
        elif isinstance(e, IndexError):
            A.update({'IndexError':i+1})
```

```python
L = [0,0,10,2,3,4,5,6]
A = {}
for i in range(len(L)):
    try:
        print(1/L[i+1])
        print(A[i])
    except Exception as e:  # Catch all exceptions
        print(e)
        if isinstance(e, ZeroDivisionError):
            A.update({'ZeroDivisionError'+str(i):L[i]})
        elif isinstance(e, IndexError):
            A.update({'IndexError'+str(i):i+1})
        if isinstance(e, KeyError):
            A.update({'KeyError'+str(i):i})
```

```python
class weather():
    '''
    This class is helping to work with weather data
    '''
    def print_function (self, x,y):
        '''
        ########### This function is print beautifully the result of the division
        ########### function
        x: list of values
        y: precision
        '''
        print('='*60)
        print(f'The values of appending {[round(w,y) for w in x]}')
        print('='*60)

    def capture_errors(self, x):
        '''
        ########### This function is catching errors
        x: list of values
        '''
        A = {}
        for i in range(len(x)):
            try:
                A.update({'value':1/x[i]})
            except Exception as e:  # Catch all exceptions
                if isinstance(e, ZeroDivisionError):
                    A.update({'ZeroDivisionError'+str(i):x[i]})

        return A

    def divison_appending_step(self, start, end, step, divisions=[2,5,3], divide=True):
        '''
        ########### This function is dividing values based on their division
        ########### to the 2 and 5

        start: number, start of values to be appended
        end: number, end of values to be appended
        divisions: list of numbers, numbers to be divided by
        step: number, number of stpes to be appended
        output: list of values

        ! hint: please be sure that step is not bigger than your length
        '''
        # empty list
        L = []
        # for loop for appending
        for x in range(start,end,step):
            if divide:
                # if condition for testing the division to 2 and 5
                if x%divisions[0]==0 or x%divisions[1]==0:
                    L.append(x)
                # if condition for testing the division to 3
                elif x%divisions[2]==0:
                    L.append(1/x)
                else:
                    L.append(2/x)
            else:
                L.append(x)

        self.print_function(L,3)
        A = self.capture_errors(L)
        print(A)
        return L

```
